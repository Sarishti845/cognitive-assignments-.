```python
#ques 1
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))

x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)

plt.plot(x, y1, label='M*x^2', color='blue', linestyle='--')
plt.plot(x, y2, label='M*sin(x)', color='red', linestyle='-')

plt.title("Plot of M*x^2 and M*sin(x)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#ques 2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Subject': ['Math', 'Physics', 'Chemistry', 'Biology', 'English'],
    'Score': [88, 75, 93, 70, 85]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
plot = sns.barplot(x='Subject', y='Score', data=df, palette='Set2')

for i, score in enumerate(df['Score']):
    plot.text(i, score + 1, str(score), ha='center')

plt.title("Scores in 5 Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
#ques 3
import numpy as np
import matplotlib.pyplot as plt

roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)

data = np.random.randn(50)

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Cumulative sum line plot
axs[0, 0].plot(np.cumsum(data), color='green')
axs[0, 0].set_title("Cumulative Sum")
axs[0, 0].set_xlabel("Index")
axs[0, 0].set_ylabel("Sum")
axs[0, 0].grid(True)

# Scatter plot with noise
axs[0, 1].scatter(range(50), data + np.random.randn(50)*0.5, color='orange')
axs[0, 1].set_title("Scatter Plot with Noise")
axs[0, 1].grid(True)

plt.tight_layout()
plt.show()
```

```python
#ques 4
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read dataset
url = 'https://github.com/AnjulaMehto/MCA/blob/main/company_sales_data.csv?raw=true'
df = pd.read_csv(url)

# 1. Total profit per month (Line plot)
plt.figure(figsize=(8, 5))
plt.plot(df['month_number'], df['total_profit'], marker='o', color='blue')
plt.title("Total Profit Each Month")
plt.xlabel("Month")
plt.ylabel("Total Profit")
plt.grid(True)
plt.show()

# 2. Product sales multiline plot
plt.figure(figsize=(10, 6))
for col in df.columns[1:-1]:  # Exclude month_number and total_profit
    plt.plot(df['month_number'], df[col], label=col)

plt.title("Monthly Sales Data")
plt.xlabel("Month")
plt.ylabel("Sales Units")
plt.legend()
plt.grid(True)
plt.show()

# 3. Bar chart for all attributes (excluding month_number)
df.drop('month_number', axis=1).sum().plot(kind='bar', figsize=(10, 6), color='teal')
plt.title("Total Sales by Product")
plt.ylabel("Units Sold")
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
