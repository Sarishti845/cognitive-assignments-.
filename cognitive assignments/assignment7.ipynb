```python
import numpy as np
import pandas as pd

roll_number = 123  # Replace this with your roll number
np.random.seed(roll_number)

sales_data = np.random.randint(1000, 5001, size=(12, 4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df = pd.DataFrame(sales_data, columns=categories, index=months)
```

```python
print(df.head())  # Display first 5 rows
print(df.describe())  # Display summary statistics

df['Total Sales'] = df.sum(axis=1)  # Total sales per month
category_totals = df.sum(axis=0)  # Total sales per category
print("Category Totals:\n", category_totals)

growth_rate = df.pct_change() * 100  # Percentage change from previous month
print(growth_rate)

if roll_number % 2 == 0:
    df['Electronics'] *= 0.9  # 10% discount
else:
    df['Clothing'] *= 0.85  # 15% discount

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df[['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']].plot(kind='line', figsize=(10, 6))
plt.title("Monthly Sales Trends")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True)
plt.show()

sns.boxplot(data=df[['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']])
plt.title("Sales Distribution by Category")
plt.show()
```

```python
array = np.array([[1, -2, 3], [-4, 5, -6]])
abs_values = np.abs(array)
print(abs_values)

percentiles = {q: np.percentile(array.flatten(), q) for q in [25, 50, 75]}
print(percentiles)

mean = np.mean(array)

median = np.median(array)
std_dev = np.std(array)
print(mean, median, std_dev)
```

```python
a = np.array([-1.8, -1.6, -0.5, 0.5, 1.6, 1.8, 3.0])
floor_values = np.floor(a)
print(floor_values)
ceil_values = np.ceil(a)
print(ceil_values)
trunc_values = np.trunc(a)
print(trunc_values)
rounded_values = np.round(a)
print(rounded_values)
```
