```python
#QUES 6
import pandas as pd

data = {
    "Employee_ID": [101, 102, 103, 104, 105],
    "Name": ["Alice", "Bob", "Charlie", "David", "Edward"],
    "Department": ["HR", "IT", "IT", "Marketing", "Sales"],
    "Age": [30, 34, 28, 35, 30],
    "Salary": [7000, 7400, 7100, 6800, 6500],
    "Years_of_Experience": [5, 8, 3, 10, 12],
    "Joining_Date": ["2020-03-15", "2017-07-19", "2019-10-04", "2021-01-10", "2010-11-25"],
    "Gender": ["Female", "Male", "Male", "Female", "Male"],
    "Bonus": [3000, 7000, 4000, 5000, 5000],
    "Rating": [4.5, 4.0, 3.8, 4.7, 3.5]
}

df = pd.DataFrame(data)
print(df)

# Shape of the DataFrame
print("Shape:", df.shape)

# Summary
print("Summary:")
print(df.info())
# Descriptive statistics
print("Descriptive Statistics:")
print(df.describe())

# Displaying first 5 and last 3 rows
print("First 5 rows:")
print(df.head(5))

print("Last 3 rows:")
print(df.tail(3))

# Calculating statistics
avg_salary = df["Salary"].mean()
total_bonus = df["Bonus"].sum()
youngest_age = df["Age"].min()
highest_rating = df["Rating"].max()

print("Average Salary:", avg_salary)
print("Total Bonus:", total_bonus)
print("Youngest Age:", youngest_age)
print("Highest Rating:", highest_rating)


# Sorting by Salary
sorted_df = df.sort_values(by="Salary", ascending=False)
print("Sorted by Salary:")
print(sorted_df)


# Categorizing employees based on Rating
def categorize_performance(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating > 4.0:
        return "Good"
    else:
        return "Average"

df["Performance_Category"] = df["Rating"].apply(categorize_performance)
print("Updated DataFrame:")
print(df)

# Checking for missing values
print("Missing Values:")
print(df.isnull().sum())

# Renaming column
df = df.rename(columns={"Employee_ID": "ID"})
print("Renamed Column:")
print(df)


# Employees with more than 5 years of experience
experienced_employees = df[df["Years_of_Experience"] > 5]
print("Employees with >5 years of experience:")
print(experienced_employees)

# Employees from IT department
it_employees = df[df["Department"] == "IT"]
print("Employees in IT department:")
print(it_employees)

# Adding Tax column (10% deduction from Salary)
df["Tax"] = df["Salary"] * 0.1
print("DataFrame with Tax column:")
print(df)

# Saving to a new CSV file
df.to_csv("modified_employees.csv", index=False)
print("Modified DataFrame saved to 'modified_employees.csv'.")
```
