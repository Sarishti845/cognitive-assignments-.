```python
#Q1

import nltk
import string
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.probability import FreqDist

nltk.download('punkt_tab')
nltk.download('stopwords')

text = """
Time travel has long captivated the human imagination, offering a tantalizing glimpse into the possibility of bending the rules of reality.
The concept suggests that one could journey backward to witness the past or leap forward to explore the future, unraveling the mysteries of history or peeking into what’s yet to come.
Email: test.email123@example.com
Visit: https://www.timetravel.com/info.html or www.example-site.org
Contact: +91 9876543210 or 123-456-7890
Pi is approximately 3.14159 and not just 3 or 3.0. The state-of-the-art machine isn’t working well.
"""

#convert text to lower case
text_lower= text.lower()
print(text_lower)
print('\n')

#remove punctuation
text_no_punct= text_lower.translate(str.maketrans('', '', string.punctuation))
print(text_no_punct)
print('\n')

#word tokenisation
words= word_tokenize(text_no_punct)
print(words)

#sentence tokenisation
sentences= sent_tokenize(text_no_punct)
print('\n')

#split using split()
words_split= text_no_punct.split()
print(words_split)
print('\n')

#split using word_tokenize()
words_word_tokenize= word_tokenize(text_no_punct)
print(words_word_tokenize)
print('\n')

#remove stop words
stop_words= set(stopwords.words('english'))
filtered_words= [word for word in words if word.casefold() not in stop_words]
print(filtered_words)
print('\n')

#word frequency distribution
freq_dist= FreqDist(filtered_words)

print("Sentences:")
for i, sent in enumerate(sentences, 1):
    print(f"{i}. {sent}")

print("\nFiltered Words:")
print(filtered_words)

print("\nWord Frequency Distribution:")
for word, freq in freq_dist.most_common():
    print(f"{word}: {freq}")
```

```python
#Q2

import re
#extract all words with only alphabets
alphabet_only_words = re.findall(r'\b[a-zA-Z]+\b', text)
print(alphabet_only_words)

#remove stop words using nltk stopword list
stop_words= set(stopwords.words('english'))
filtered_words= [word for word in words if word.casefold() not in stop_words]
print(filtered_words)
print('\n')

#applying stemming to tokenised words in Q1
from nltk.stem import PorterStemmer, LancasterStemmer

porter_stemmer= PorterStemmer()
lancaster_stemmer= LancasterStemmer()

#stemmed words
porter_stemmed_words= [porter_stemmer.stem(word) for word in filtered_words]
lancaster_stemmed_words= [lancaster_stemmer.stem(word) for word in filtered_words]

print("Porter Stemmed Words:")
print(porter_stemmed_words)

print("\nLancaster Stemmed Words:")
print(lancaster_stemmed_words)

#lemmatization
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')

lemmatizer= WordNetLemmatizer()
lemmatized_words= [lemmatizer.lemmatize(word) for word in filtered_words]

print("\nLemmatized Words:")
print(lemmatized_words)
```

```python
#q4

import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Define texts
text_1 = "Artificial Intelligence (AI) is the simulation of human intelligence processes by machines. It involves learning, reasoning, and self-correction. AI technologies are used in a variety of fields such as robotics, language processing, and automation. Machine learning and deep learning are subfields of AI that enable systems to learn from data and improve over time."
text_2 = "Blockchain is a decentralized digital ledger technology that records transactions across many computers. It ensures transparency, security, and immutability of data. Blockchain is most commonly associated with cryptocurrencies like Bitcoin but can be applied in various industries such as finance, supply chain management, and healthcare. Its structure allows for secure, transparent, and distributed data management."

# Preprocess and tokenize
def preprocess(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^a-z\s]', '', text)  # Remove punctuation and non-alphabetic characters
    tokens = text.split()  # Tokenize by space
    return set(tokens)

tokens_1 = preprocess(text_1)
tokens_2 = preprocess(text_2)

print("Tokens from Text 1:", tokens_1)
print("Tokens from Text 2:", tokens_2)

# Calculate Jaccard Similarity
intersection = tokens_1.intersection(tokens_2)
union = tokens_1.union(tokens_2)

jaccard_similarity = len(intersection) / len(union)
print(f"Jaccard Similarity: {jaccard_similarity:.4f}")

# Tfidf Vectorizer
texts = [text_1, text_2]
vectorizer = TfidfVectorizer()

# Fit and transform the texts into TF-IDF matrix
tfidf_matrix = vectorizer.fit_transform(texts)

# Calculate cosine similarity
cosine_sim = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])
print(f"Cosine Similarity: {cosine_sim[0][0]:.4f}")
```

```python
#q5

from textblob import TextBlob

# Review text
review_text = "I recently bought the XYZ smartphone, and I am really impressed by its performance. The battery life is incredible, and the camera quality is fantastic. The phone is sleek, fast, and easy to use. I highly recommend it to anyone looking for a new phone!"

# Create a TextBlob object
blob = TextBlob(review_text)

# Get polarity and subjectivity
polarity = blob.sentiment.polarity
subjectivity = blob.sentiment.subjectivity

# Classify the review based on polarity
if polarity > 0:
    sentiment = "Positive"
elif polarity < 0:
    sentiment = "Negative"
else:
    sentiment = "Neutral"

# Print results
print(f"Polarity: {polarity}")
print(f"Subjectivity: {subjectivity}")
print(f"Sentiment: {sentiment}")

from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Create a list of positive reviews
positive_reviews = [
    "I recently bought the XYZ smartphone, and I am really impressed by its performance.",
    "The battery life is incredible, and the camera quality is fantastic.",
    "The phone is sleek, fast, and easy to use. I highly recommend it to anyone looking for a new phone!",
    "I love the design and speed of the XYZ smartphone. It's a great value for money!"
]

# Join all positive reviews into one string
positive_text = " ".join(positive_reviews)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(positive_text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

```

```python
#Q6

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np

# Define the input text
text = "Machine learning is fun and exciting to learn"

# Initialize the tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts([text])

# Create sequences of tokens
sequences = []
words = text.split()

# Generate sequences of words (sliding window approach)
for i in range(1, len(words)):
    seq = words[:i+1]  # Get the first i+1 words
    tokenized_seq = tokenizer.texts_to_sequences([' '.join(seq)])[0]  # Convert to sequence of tokens
    sequences.append(tokenized_seq)

# Pad the sequences to ensure uniform length
padded = pad_sequences(sequences)

# Prepare the data for training (input and labels)
X_input = padded[:, :-1]  # All columns except the last one
y_input = padded[:, -1]   # Only the last column as the target label

print("Input Sequences (X):")
print(X_input)
print("Target Sequences (y):")
print(y_input)

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense

# Define the model
model = Sequential()

# Embedding layer for input sequences
model.add(Embedding(input_dim=len(tokenizer.word_index) + 1,  # Vocabulary size
                    output_dim=10,                          # Embedding dimension
                    input_length=X_input.shape[1]))         # Input length is the sequence length

# LSTM layer to capture sequential dependencies
model.add(LSTM(50))

# Fully connected dense layer
model.add(Dense(50, activation='relu'))

# Output layer with softmax for classification
model.add(Dense(len(tokenizer.word_index) + 1, activation='softmax'))  # Output layer size is vocab size

# Compile the model
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam')

# Print model summary
model.summary()

# Train the model (you would normally use a larger dataset for more epochs)
model.fit(X_input, y_input, epochs=200, verbose=1)

# Function to generate text based on a seed word
def generate_text(seed_text, model, tokenizer, max_sequence_length, num_words=10):
    for _ in range(num_words):
        # Convert seed text to sequence
        sequence = tokenizer.texts_to_sequences([seed_text])[0]
        sequence = pad_sequences([sequence], maxlen=max_sequence_length - 1, padding='pre')

        # Predict next word
        predicted_word_index = np.argmax(model.predict(sequence, verbose=0), axis=-1)[0]
        predicted_word = tokenizer.index_word[predicted_word_index]

        # Append the predicted word to the seed text
        seed_text += " " + predicted_word

    return seed_text

# Generate new text from a seed
seed_text = "Machine learning"
generated_text = generate_text(seed_text, model, tokenizer, X_input.shape[1])
print("Generated Text:\n", generated_text)
```
